## Function to generate gaussian curves with various amplitudes

```python
# Function to generate Gaussian curve
def gaussian(x, amplitude, center, stddev):
    return amplitude * np.exp(-((x - center) / stddev) ** 2)

# Parameters
amplitudes = [1, 2, 3, 4, 5, 6, 7, 6, 5, -4, -3]
center = 0
stddev = 1

# Generate x values
x = np.linspace(-10, 10, 100)

# Generate and plot Gaussian curves
for amplitude in amplitudes:
    y = gaussian(x, amplitude, center, stddev)
    plt.plot(x, y)

# Set plot labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gaussian Curves')

# Display the plot
plt.show()
```

```python
def plot_csd_heatmap(data_matrix):
    # Define layer names
    layer_names = ['Layer 1', 'Layer 2', 'Layer 3', 'Layer 4','Layer 5']

    # Compute the CSD
    csd = np.gradient(data_matrix, axis=0, edge_order=2)

    # Create a time axis
    time = np.arange(data_matrix.shape[1])

    # Create a depth axis
    depth = np.arange(data_matrix.shape[0])

    # Create a grid for plotting
    time_grid, depth_grid = np.meshgrid(time, depth)

    # Plot the CSD as a heatmap
    plt.figure(figsize=(8, 4))
    plt.imshow(csd, cmap="RdBu_r", aspect='auto', extent=[time[0], time[-1], depth[0], depth[-1]])
    plt.colorbar(label='Current Source Density')
    plt.xlabel('Time')
    plt.ylabel('Layer')
    plt.yticks(depth, layer_names)  # Set y-axis tick labels
    plt.title('CSD of generated signal')
    plt.show()
```

# smoothing the data

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_with_noise(x, amplitude, center=0, stddev=1, noise_stddev=0.5):
    noise = np.random.normal(0, noise_stddev, len(x))
    return amplitude * np.exp(-((x - center) / stddev) ** 2) + noise

def moving_average(data, window_size):
    cumsum = np.cumsum(data)
    cumsum[window_size:] = cumsum[window_size:] - cumsum[:-window_size]
    return cumsum[window_size - 1:] / window_size

def plot_gaussian_curves_with_noise(amplitudes, center=0, stddev=1, noise_stddev=0.5, smoothing_window=10):
    num_layers = len(amplitudes)

    # Generate x values
    x = np.linspace(-10, 10, 1000)

    # Initialize matrix to store generated data
    data_matrix = np.zeros((num_layers, len(x)))

    # Generate and plot Gaussian curves with noise
    for i, amplitude in enumerate(amplitudes):
        y = gaussian_with_noise(x, amplitude, center, stddev, noise_stddev)
        data_matrix[i] = y
        plt.plot(x, y, label=f'Layer {i+1}')

    # Set plot labels and title
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Gaussian Curves with Noise')
    plt.legend()

    # Display the plot
    plt.show()

    # Apply smoothing to the data
    smoothed_data_matrix = np.apply_along_axis(moving_average, 1, data_matrix, smoothing_window)

    # Plot CSD heatmap
    layer_names = [f'Layer {i+1}' for i in range(num_layers)]
    csd = np.gradient(smoothed_data_matrix, axis=0, edge_order=2)
    csd = np.gradient(csd, axis=0, edge_order=1)
    time = np.arange(smoothed_data_matrix.shape[1])
    depth = np.arange(smoothed_data_matrix.shape[0])
    time_grid, depth_grid = np.meshgrid(time, depth)

    plt.figure(figsize=(8, 4))
    plt.imshow(csd, cmap="RdBu_r", aspect='auto', extent=[time[0], time[-1], depth[0], depth[-1]])
    plt.colorbar(label='Current Source Density')
    plt.xlabel('Time')
    plt.ylabel('Layer')
    plt.yticks(depth, layer_names)
    plt.title('CSD of generated signal with smoothing')
    plt.show()

# Parameters
amplitudes = [3, 9, 0.1, -7, -5]

# Call the function to plot Gaussian curves with noise and CSD heatmap with smoothing
plot_gaussian_curves_with_noise(amplitudes, smoothing_window= 1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_with_noise(x, amplitude, center=0, stddev=1, noise_stddev=0.5):
    noise = np.random.normal(0, noise_stddev, len(x))
    return amplitude * np.exp(-((x - center) / stddev) ** 2) + noise

def plot_gaussian_curves_with_noise(amplitudes, center=0, stddev=1, noise_stddev=0.5, smoothing_sigma=2):
    num_layers = len(amplitudes)

    # Generate x values
    x = np.linspace(-10, 10, 1000)

    # Initialize matrix to store generated data
    data_matrix = np.zeros((num_layers, len(x)))

    # Generate and plot Gaussian curves with noise
    for i, amplitude in enumerate(amplitudes):
        y = gaussian_with_noise(x, amplitude, center, stddev, noise_stddev)
        data_matrix[i] = y
        plt.plot(x, y, label=f'Layer {i+1}')

    # Set plot labels and title
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Gaussian Curves with Noise')
    plt.legend()

    # Display the plot
    plt.show()
    
    data_matrix = gaussian_filter1d(data_matrix, sigma=smoothing_sigma, axis=1)

    # Compute the spatial second-order derivative along the depth axis manually
    csd_depth = np.zeros_like(data_matrix)

    for i in range(1, num_layers - 1):
        csd_depth[i] = data_matrix[i + 1] - 2 * data_matrix[i] + data_matrix[i - 1]

    csd_depth_second_derivative = np.zeros_like(data_matrix)
    for i in range(1, num_layers - 1):
        csd_depth_second_derivative[i] = csd_depth[i + 1] - 2 * csd_depth[i] + csd_depth[i - 1]

    # Plot CSD heatmap
    layer_names = [f'Layer {i+1}' for i in range(num_layers)]
    depth = np.arange(data_matrix.shape[0])

    plt.figure(figsize=(8, 4))
    plt.imshow(csd_depth_second_derivative, cmap="RdBu_r", aspect='auto', extent=[x[0], x[-1], depth[0], depth[-1]])
    plt.colorbar(label='Second-Order Derivative (Depth axis)')
    plt.xlabel('Time')
    plt.ylabel('Layer')
    plt.yticks(depth, layer_names)
    plt.title('Second-Order Derivative along Depth axis')
    plt.show()

# amplitudes = [3, 9, 0.1, -7, -5]
# plot_gaussian_curves_with_noise(amplitudes, smoothing_sigma=10)

# amplitudes = [x for x in range(1, 11, 2)]
# print(len(amplitudes))
# plot_gaussian_curves_with_noise(amplitudes, smoothing_sigma=10)

amplitudes = [3 ** x for x in range(5)]  # 2^0 to 2^5 = 1 to 32
print(amplitudes)
plot_gaussian_curves_with_noise(amplitudes, smoothing_sigma=10)
```
