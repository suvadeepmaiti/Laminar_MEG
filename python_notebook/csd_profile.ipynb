```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to generate Gaussian curve
def gaussian(x, amplitude, center, stddev):
    return amplitude * np.exp(-((x - center) / stddev) ** 2)

# Parameters
amplitudes = [1, 2, 3, 4, 5, 6, 7, 6, 5, -4, -3]
center = 0
stddev = 1

# Generate x values
x = np.linspace(-10, 10, 100)

# Generate and plot Gaussian curves
for amplitude in amplitudes:
    y = gaussian(x, amplitude, center, stddev)
    plt.plot(x, y)

# Set plot labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gaussian Curves')

# Display the plot
plt.show()
```

```python
# Function to generate Gaussian curve with noise
def gaussian_with_noise(x, amplitude, center, stddev, noise_stddev):
    noise = np.random.normal(0, noise_stddev, len(x))
    return amplitude * np.exp(-((x - center) / stddev) ** 2) + noise

# Parameters
amplitudes = [0.5, 0.1, 7, 4, 0.8, 6, 7, 6, 5, 4, 3]
center = 0
stddev = 1
noise_stddev = 0.5

# Generate x values
x = np.linspace(-10, 10, 1000)

# Initialize matrix to store generated data
data_matrix = np.zeros((len(amplitudes), len(x)))

# Generate and plot Gaussian curves with noise
for i, amplitude in enumerate(amplitudes):
    y = gaussian_with_noise(x, amplitude, center, stddev, noise_stddev)
    data_matrix[i] = y
    plt.plot(x, y)

# Set plot labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gaussian Curves with Noise')

# Display the plot
plt.show()

# Print the data matrix
print("Data Matrix:")
print(data_matrix)
```

```python
def plot_csd_heatmap(data_matrix, layer):
    # Define layer names
    layer_names = ['Layer 1', 'Layer 2', 'Layer 3', 'Layer 4', 'Layer 5', 'Layer 6', 'Layer 7', 'Layer 8', 'Layer 9', 'Layer 10', 'Layer 11']

    # Compute the CSD
    csd = np.gradient(data_matrix, axis=0)

    # Create a time axis
    time = np.arange(data_matrix.shape[1])

    # Create a depth axis
    depth = np.arange(data_matrix.shape[0])

    # Create a grid for plotting
    time_grid, depth_grid = np.meshgrid(time, depth)

    # Plot the CSD as a heatmap
    plt.figure(figsize=(8, 4))
    plt.imshow(csd, cmap="RdBu_r", aspect='auto', extent=[time[0], time[-1], depth[0], depth[-1]])
    plt.colorbar(label='Current Source Density')
    plt.xlabel('Time')
    plt.ylabel('Layer')
    plt.yticks(depth, layer_names)  # Set y-axis tick labels
    plt.title('CSD of source at layer ' + str(layer))
    plt.show()
```
