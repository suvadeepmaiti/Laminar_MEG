```python
import numpy as np
import scipy
import scipy.io
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from scipy.ndimage import gaussian_filter
```

```python
def plot_csd_heatmap(csd_data, layer):
    # Load the CSD data from the provided file
    mat_data = scipy.io.loadmat(csd_data)
    source_tcs = mat_data['source_tcs']
    data = source_tcs

    # Define layer names
    layer_names = ['Layer 1', 'Layer 2', 'Layer 3', 'Layer 4', 'Layer 5', 'Layer 6', 'Layer 7', 'Layer 8', 'Layer 9', 'Layer 10', 'Layer 11']

    # Compute the CSD
    csd = np.gradient(data, axis=0)

    # Create a time axis
    time = np.arange(data.shape[1])

    # Create a depth axis
    depth = np.arange(data.shape[0])

    # Create a grid for plotting
    time_grid, depth_grid = np.meshgrid(time, depth)

    # Plot the CSD as a heatmap
    plt.figure(figsize=(8, 4))
    plt.imshow(csd, cmap="RdBu_r", aspect='auto', extent=[time[0], time[-1], depth[0], depth[-1]])
    plt.colorbar(label='Current Source Density')
    plt.xlabel('Time')
    plt.ylabel('Layer')
    plt.yticks(depth, layer_names)  # Set y-axis tick labels
    plt.title('CSD of source at index 18527 layer '+ str(layer))
    plt.show()
```

```python
csd_data3 = '/home/smaiti/Danc_lab/Final_Experiments/csd_processing_files/sim@index_18527_layer_3to9/sim@index_18527_layer_3.mat'
plot_csd_heatmap(csd_data3, 3)
```
